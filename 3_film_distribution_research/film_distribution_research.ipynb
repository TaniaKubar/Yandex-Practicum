# Исследование рынка российского кинопроката

Заказчик этого исследования — Министерство культуры Российской Федерации.
Требуется изучить рынок российского кинопроката и выявить текущие тренды. А также отдельно рассмотреть фильмы, которые получили государственную поддержку, и ответить на вопрос, насколько такие фильмы интересны зрителю.
Данные опубликованны на портале открытых данных Министерства культуры. Набор данных содержит информацию о прокатных удостоверениях, сборах и государственной поддержке фильмов, а также информацию с сайта КиноПоиск.


**Цель исследования** — выявить текущие тренды рынка российского кинопроката, оценить популярность и окупаемость фильмов, получивших государственную поддержку.

Затем нужно добавить колонки с данными, необходимыми для дальнейшего анализа.

Выведем на экран первые, последние и случайные пять строк таблицы `df`, а также общую информацию. Создадим для этого собственную функцию:

```python
def total_view(df: pd.DataFrame, n_rows=5, seed=None):
    '''Возвращает первые, случайные и последние строки таблицы (по умолчанию по 5 строк)'''
    return display(pd.concat([
                              df.head(n_rows),
                              df.sample(n_rows, random_state=seed),
                              df.tail(n_rows)
                             ]))

total_view(df)
display(df.describe().T)
df.info()
```

## Слияние датасетов <a id='merge'></a>

Перед слиянием необходимо привести тип даных в `df['puNumber']` к числовому типу, некорректные значения заменим на NaN:

```python
df['puNumber'] = pd.to_numeric(df['puNumber'], errors='coerce')
df = df.dropna(subset=['puNumber'])
```

Объединим данные в один датафрейм по колонке `puNumber` таким образом, чтобы все объекты из `df` обязательно вошли в получившийся датафрейм:

```python
df = df.merge(df_shows, on='puNumber', how='left')
total_view(df)
```

```python
df.info()
```

Таблицы 'склеились', общее количество строк не изменилось. Добавилась колонка `box_office`, тип данных `float64`, заполнена примерно на 42%.

```python
df = df.rename(columns={'puNumber': 'pu_number'})
df.columns
```

### Типы данных <a id='types'></a>


Тип данных в `pu_number` изменен, остались колонки `show_start_date` и `ratings`. В колонке `show_start_date` указаны даты премьеры фильмов, изменим тип данных на `datetime`:

```python
df['show_start_date'] = pd.to_datetime(df['show_start_date'], format='%Y-%m-%dT%I:%M:%S.%fZ')
df['show_start_date'][:10]
```

Посмотрим на значения в колонке `ratings`:

```python
df['ratings'].unique()
```

В колонке хранятся дробные числа и проценты, тип данных `object`. Чтобы изменить тип данных на `float` и привести их к одной системе оценки, необходимо убрать знаки `%` и разделить соответствующие числа на 10:

```python
df['ratings'] = df['ratings'].str.replace('%', '').astype('float')
df.loc[df['ratings'] > 10, 'ratings'] = df.loc[df['ratings'] > 10, 'ratings'] / 10
df['ratings'].unique()
```

В колонках `show_start_date` `refundable_support`, `nonrefundable_support`, `budget` и `box_office` хранятся количественные значения. Рассмотрим их подробнее, начнем с колонки `show_start_date`:

Переведем суммы в колонках с данными о финансировании и кассовых сборах в миллионы для удобства:

```python
df['refundable_support'] = round(df['refundable_support'] / 1_000_000, 2)
df['nonrefundable_support'] = df['nonrefundable_support'] / 1_000_000
df['budget'] = df['budget'] / 1_000_000
df['box_office'] = df['box_office'] / 1_000_000
```

Построим диаграммы размаха по данным в колонках `refundable_support`, `nonrefundable_support`, `budget` и `box_office`, чтобы определить выбивающиеся значения:

```python
column_list = ['refundable_support', 'nonrefundable_support', 'budget', 'box_office']
cnt = 1
plt.figure(figsize=(15, 15))
for column in column_list:
    plt.subplot(2, 2, cnt)
    df[df[column].notna()].boxplot(column)
    plt.ylabel('млн. руб.')
    if cnt == 1:
        plt.title('Возвратные средства гос. поддержки')
    if cnt == 2:
        plt.title('Невозвратные средства гос. поддержки')
    if cnt == 3:
        plt.title('Общий бюджет фильма')
    if cnt == 4:
        plt.title('Кассовые сборы в России')
    cnt += 1
```

## Добавление расчетных данных <a id='newdata'></a>
### Год проката <a id='year'></a>
Добавим колонку с годом проката:

```python
df['year'] = pd.DatetimeIndex(df['show_start_date']).year
df['year'][:10]
```

### Главный режиссер <a id='director'></a>
Добавим колонку с именем и фамилией главного режиссера (первое значение из списка режиссеров по каждой записи). Фамилии режиссеров разделены запятыми, напишем собственную функцию с учетом разделителя и применим ее к колонке `director`:

```python
# Собственная функция, возвращающая нужную подстроку из строки
def main_value(value):
    '''Возвращает подстроку, используя правило:
    
    если передается тип данных str, делит строку на элементы по разделителю ','
    и возвращает первый элемент полученного списка
    
    '''
    if type(value) == str:
        new_value = value.split(',')
        return new_value[0]
# Создание колонки 'main_director' и запись в нее результата работы функции main_value
df['main_director'] = df['director'].apply(main_value)
df['main_director'][:10]
```

### Жанр <a id='genre'></a>
Добавим колонку с основным жанром фильма (первое значение из списка жанров по каждой записи). Жанры, как и режиссеры, разделены запятыми. Используем для этого собственную функцию `main_value`:

```python
df['main_genre'] = df['genres'].apply(main_value)
df['main_genre'][:10]
```

### Доля государственной поддержки в бюджете фильма <a id='ratio'></a>
Добавим колонку с долей государственной поддержки от общего бюджета фильма в процентах:

```python
df['support_ratio'] = round(100 * (df['refundable_support'] + df['nonrefundable_support']) / df['budget'], 2)
df.loc[df['support_ratio'] > 0, 'support_ratio'][:10]
```

### Суммарная государственная поддержка и окупаемость фильма <a id='support'></a>
Добавим колонки с суммарной государственной поддержкой и окупаемостью фильма в процентах:

```python
df['sum_support'] = round(df['refundable_support'] + df['nonrefundable_support'], 2)
df['payback'] = round(100 * df['box_office'] / df['budget'], 2)
df.loc[df['sum_support'] >0, ['sum_support', 'payback']][:10]
```

### Наличие  государственной поддержки<a id='is_support'></a>
Добавим колонку с булевыми значениями, где `True` соответствует фильмам, получившим государственную поддержку, а `False` не получившим:
